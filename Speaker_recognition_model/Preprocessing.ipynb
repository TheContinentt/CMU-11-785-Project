```python
from pydub import AudioSegment
import csv
```

```python
def read_txt(file_path):
    res = []
    f = open(file_path, "r")
    lines = f.readlines()
    for line in lines:
      res.append(line.strip())
    return res
```

```python
train_file = '/content/gdrive/MyDrive/knnw/train.txt'
# val_file = '/content/drive/MyDrive/knnw/val.txt'
test_file = '/content/gdrive/MyDrive/knnw/test.txt'

train_idx = read_txt(train_file)
# val_idx = read_txt(val_file)
test_idx = read_txt(test_file)
print(len(train_idx), len(test_idx))
```

```python
def read_id(file_path):
    res = {}
    label_set = set()
    f = open(file_path, "r")
    lines = f.readlines()
    for line in lines:
      curr_line = line.strip().split(',')
      label_set.add(curr_line[1])
      res[curr_line[0]] = curr_line[1]
    return res, label_set
```

```python
id_speaker = '/content/gdrive/MyDrive/knnw/id_speaker.txt'
id_speaker_dict, label_set = read_id(id_speaker)
print(id_speaker_dict, label_set)
```

```python
label_to_int = dict()
idx = 0
for label in label_set:
  label_to_int[label] = idx
  idx += 1
```

```python
import os

for label in label_set:
  os.mkdir('/content/gdrive/MyDrive/knnw/data/test/' + str(label_to_int[label]) + '/')
  os.mkdir('/content/gdrive/MyDrive/knnw/data/train/' + str(label_to_int[label]) + '/')
```

```python
csv_file = '/content/gdrive/MyDrive/knnw/knnw_en_sub.csv'
wav_file = '/content/gdrive/MyDrive/knnw/knnw_en_mono.wav'
orig_wav = AudioSegment.from_wav(wav_file)
train_list = []
test_list = []
with open(csv_file, mode='r') as f:
    csv_reader = csv.reader(f, delimiter=';')
    next(csv_reader)
    for row in csv_reader:
        id, start_time, end_time, text = row
        if id not in id_speaker_dict:
          continue
        if id in train_idx:
          wav_file = '/content/gdrive/MyDrive/knnw/data/train/'+str(label_to_int[id_speaker_dict[id]])+'/'+str(id)+'.wav'
          train_list.append((wav_file, label_to_int[id_speaker_dict[id]]))
        else:
          wav_file = '/content/gdrive/MyDrive/knnw/data/test/'+str(label_to_int[id_speaker_dict[id]])+'/'+str(id)+'.wav'
          test_list.append((wav_file, label_to_int[id_speaker_dict[id]]))
        new_wav = orig_wav[int(start_time):int(end_time)]
        new_wav.export(wav_file, format="wav")
```

```python
print(len(train_list), len(test_list), len(train_list + test_list), train_list[0])
```

```python
import numpy as np
data = np.load('TIMIT_labels.npy', allow_pickle=True)
print(data)
all_list = train_list + test_list
all_scp = {}
for elem in all_list:
  all_scp[elem[0]] = elem[1]
np.save('new_TIMIT_labels.npy', all_scp)
```

```python
train_scp_list = []
for elem in train_list:
  train_scp_list.append(elem[0])
test_scp_list = []
for elem in test_list:
  test_scp_list.append(elem[0])
all_scp_list = []
for elem in all_list:
  all_scp_list.append(elem[0])
```

```python
with open('TIMIT_train.scp', 'w+') as f:
  for elem in train_scp_list:
    f.write('%s\n' %elem)
f.close()

with open('TIMIT_test.scp', 'w+') as f:
  for elem in test_scp_list:
    f.write('%s\n' %elem)
f.close()

with open('TIMIT_all.scp', 'w+') as f:
  for elem in all_scp_list:
    f.write('%s\n' %elem)
f.close()
```
